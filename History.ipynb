###3 Vs of Big Data

📊 1. Volume
Refers to the amount of data generated and stored.
Big data involves massive datasets—from terabytes to petabytes and beyond.
Examples: Social media posts, sensor data, transaction logs.

⚡ 2. Velocity
Describes the speed at which data is generated, processed, and analyzed.
Real-time or near-real-time data streams are common in big data environments.
Examples: Stock market feeds, live video streams, IoT sensor updates.

🧩 3. Variety
Indicates the different types and sources of data.
Data can be structured (tables), semi-structured (XML, JSON), or unstructured (images, videos, text).
Examples: Emails, tweets, medical records, satellite imagery

###**Data** **Lakehouse**

A **Lakehouse** is a modern data architecture that combines the best features of both data lakes and data warehouses into a single, unified platform.

🔑 **Key** **Benefits**
1. Unified storage: No need for separate systems for raw and structured data
2. Cost-effective: Uses low-cost cloud object storage (e.g., Amazon S3, Azure Data Lake)
3. Real-time analytics: Supports both batch and streaming data ingestion
4. ACID transactions: Ensures data integrity with technologies like **Delta Lake** or Apache Iceberg
5. Governance and security: Centralized metadata and access control

**Delta Lake** is an open-source storage layer that brings reliability, performance, and advanced features to data lakes—transforming them into lakehouses. It was originally developed by Databricks and is tightly integrated with Apache Spark.

🧠 **Key Features of Delta Lake**
1. ACID Transactions: Ensures data integrity with atomicity, consistency, isolation, and durability—even in big data environments.
2. Schema Enforcement & Evolution: Prevents schema mismatches and allows gradual updates to data structure.
3. Time Travel: Lets you query previous versions of your data using built-in versioning.
4. Scalable Metadata Handling: Improves performance for large-scale datasets.
5. Unified Batch & Streaming: Supports both real-time and batch data processing using a single copy of data.

🗂️ **How It Works**

Delta Lake stores data in Parquet files and adds a transaction log (called the DeltaLog) to track changes. This log enables features like rollback, audit trails, and version control.

###Apache Spark

**Apache Spark** is an open-source, distributed computing system designed for fast and scalable data processing. It’s widely used for big data analytics, machine learning, and real-time stream processing.

⚡ **Key Features**

1. **In-memory computing:** Spark processes data in RAM, making it much faster than traditional disk-based engines like Hadoop MapReduce.
2. **Unified engine:** Supports multiple workloads—batch processing, interactive queries, streaming data, and ML — all in one platform.
3. **Scalability:** Can run on a single machine or scale across thousands of nodes in a cluster.
4. **Rich APIs:** Offers high-level APIs in Python, Java, Scala, and R, making it accessible to a wide range of developers.

🧠 **Core Components**

1. Spark Core: The foundation for distributed task scheduling and memory management.
2. Spark SQL: Enables querying structured data using SQL or DataFrame APIs.
3. Spark Streaming: Processes real-time data streams.
4. MLlib: Built-in machine learning library for scalable algorithms.
5. GraphX: For graph processing and analytics.

###Databricks Cluster

In Databricks, a **cluster** is a set of computing resources used to run data engineering, data science, and machine learning workloads. It’s the backbone of processing in the Databricks environment.

🧠 **What Does a Cluster Do?**

1. Executes notebooks, jobs, and queries
2. Distributes tasks across multiple nodes for parallel processing
3. Provides the environment for Apache Spark to run

🧱 **Cluster Components**
1. **Driver node:** Coordinates the execution of tasks and maintains SparkContext.
2. **Worker nodes:** Execute the tasks assigned by the driver and store data in memory or disk.

⚙️ **Types of Clusters in Databricks**

1.  **Interactive clusters:** Used for development and exploration in notebooks.
2.  **Job clusters:** Automatically spun up for scheduled jobs and terminated after completion.
3.  **All-purpose clusters:** Shared among users for collaborative work.

🔑  **Key Features**

1. **Auto-scaling:** Automatically adjusts the number of worker nodes based on workload.
2. **Auto-termination:** Shuts down idle clusters to save costs.
3. **Cluster pools:** Pre-warmed resources to reduce startup time.

###Databricks Architecture

![](https://learn.microsoft.com/en-us/azure/databricks/_static/images/getting-started/architecture-azure.png)

🏗️ **Core Architectural Layers**

1. **Control Plane**	Managed by Databricks. Handles authentication, job scheduling, UI, and metadata.
2. **Compute Plane**	Where Spark clusters run. Can be serverless or customer-managed. Processes data.
3. **Storage Layer**	Uses cloud object storage (e.g., AWS S3, Azure Data Lake, GCS) for storing data.

Compute Plane is divided into Serverless and Classic Computes

☁️ **Serverless Compute**

1. A fully managed compute environment where the platform automatically provisions, scales, and terminates resources based on workload demand—without requiring user configuration or cluster management.
2. You don’t need to set up or manage clusters.
3. Ideal for quick, cost-efficient, and scalable execution of jobs and queries.
4. VMs will be created in Databricks and it resides in Databricks account

🖥️ **Classic Compute**

1. A user-managed compute environment where you manually configure clusters, control scaling, and manage lifecycle settings to suit specific workload requirements.
2. Offers full control over cluster size, type, and behavior.
3. Best for complex, long-running, or highly customized workloads.
4. VMs will be created in Cloud (Azure) Account and it resides in Cloud account

###Medallion Architecture

**Bronze Layer:**
1. Ingests raw data from sources like Kafka, APIs, or cloud storage.
2. Stores it in Delta Lake format for reliability.
3. Minimal processing—preserves original fidelity.

**Silver Layer:**
1. Applies transformations: filtering, joins, deduplication, schema enforcement.
2. Produces clean, structured datasets for downstream use.

**Gold Layer:**
1. Aggregates and summarizes data for business intelligence, dashboards, and machine learning.
2. Often used by analysts and decision-makers.

🚀 **Benefits**\
**Modularity**: Each layer can be independently managed and optimized.\
**Data Quality**: Ensures clean and trusted data flows.\
**Scalability**: Supports massive datasets with Delta Lake and Apache Spark.\
**Reusability**: Silver and Gold datasets can serve multiple use cases.
