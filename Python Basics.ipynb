```python
print("Hello Guru", 1, "LPU")            #Hello Guru 1 LPU
print("Hello Guru", 1, "LPU", sep=" ")   #Hello Guru 1 LPU   -- Space is default Separator
print("Hello Guru", 1, "LPU", sep=":")   #Hello Guru:1:LPU
print("""Hello, How are you?
      How is your day going?""")         #New Line use ''' or """
print("Hello 'Manju' How are you?")      #Hello 'Manju' How are you?
print('Hello "Manju" How are you?')      #Hello "Manju" How are you?
print("Hello \'Manju\' How are you?")    #Hello 'Manju' How are you? Escapes or Ignores '
```
